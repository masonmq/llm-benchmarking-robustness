### Keep only the two columns I need

```python
df = df_full[['complaints_2008','first_A']]
```

```python
df
```

```python
# first_A of 0 means firm name does NOT begin with A or a number
df['first_A'].value_counts()
```

```python
# Groupby name type, calculate summary metrics
df.groupby('first_A').agg(['count','mean','std','min','max'])
```

```python
# Save groupby df
grouped_df = df.groupby('first_A').agg(['count','mean','std','min','max'])
```

```python
grouped_df['complaints_2008']['mean']
```

```python
# How much higher is the mean number of complaints for first_A = 1 (those beginning with A or number)?
# 1.33 / .248
grouped_df['complaints_2008']['mean'][1] / grouped_df['complaints_2008']['mean'][0]
```
